```python
# Import Library
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open the Data set
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
# Remove the categorical data(i.e str-state)
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
# splite the input & output columns
```

```python
dataset.columns
```

```python
independant=dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
independant
```

```python
dependant= dataset[['charges']]
```

```python
dependant
```

```python
#split the Train & Test dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
# Train data-Model creation
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid= { 'criterion':['squared_error','absolute-error','mae','friedman_mse','poisson'],'max_features':['auto','sqrt','log2'],'splitter':['best','ramdom']}

grid= GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)

# Fitting the Model for grid seach                 
grid.fit(X_train,Y_train)
```

```python
# print best parameter after turning
# print(grid_best_params_)
```

```python
re=grid.cv_results_
print("The r_score_value for best parameters:",format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
