```python
#Import Library
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open the dataset
```

```python
dataset= pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#Remove the catagorical data (Converting str to Numbers)
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
#Splite input & Out put data set
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
# split the Train set & test set dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
# Model Improve for using the Standardisation
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_test
```

```python
# Model creation-Train set
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid= {'kernel':['rbf','ploy','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auti','scale']}
grid= GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)

# Fitting the Model for grid seach                 
grid.fit(X_train,Y_train)
                 
```

```python
# print best parameter after turning
# print(grid_best_params_)
```

```python
re=grid.cv_results_
print("The r_score_value for best parameters:",format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
